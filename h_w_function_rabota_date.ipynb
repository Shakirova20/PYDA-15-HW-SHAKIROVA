```python
import pandas as pd
```

```python
data = pd.read_csv('movies.csv')                                            
data.head()
```

```python
data_ = pd.read_csv('ratings.csv')
data_.head()
```

```python
df = data.merge(data_, on='movieId').head()
df.head()
```

```python
def class_(param):
    if param <= 2:
        return 'низкий рейтинг'
    elif param <= 4:
        return 'средний рейтинг'
    elif param >= 4.5 and param <=5:
        return 'высокий рейтинг'

df['class_'] = df['rating'].apply(class_)
df.tail()
```

```python
df_r = pd.read_csv('keywords.csv')
df_r.head()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

city = {value: key for key in geo_data for value in geo_data[key]}
print(city)

df_r = pd.read_csv('keywords.csv')

def region(value):
    words = [word.strip().lower() for word in value.split(' ')]
    for word in words:
        if word in city:
            return city.get(word)
    return 'undefined'

df_r['region'] = df_r['keyword'].apply(region)
df_r.head(10)
```

```python
df_r.region.value_counts()
```

```python
df_sort = (df_r[ (df_r['region']=='Дальний Восток')])
df_sort.head()
```

```python
display(df_r.groupby(by='region').head(5))
```
